## 자동

- ㅎㅎ 웹크롤링 코드 짤때는 머리 아픈데 막상 완성하고 원하는 방식대로 돌아갈때면 기분이 좋다

```python
import time

now = time

driver = webdriver.Chrome()
driver.get("http://")
driver.implicitly_wait(3)

for t in range(1,20):
    
    driver.refresh()
    
    driver.find_element_by_xpath('//*[@id="sch_sido"]').click()
    driver.find_element_by_xpath('//*[@id="sch_sido"]/option[10]').click()
    driver.find_element_by_xpath('//*[@id="sch_gungu"]').click()
    driver.find_element_by_xpath('//*[@id="sch_gungu"]/option[24]').click()
    driver.find_element_by_xpath('//*[@id="sch_stat"]').click()
    driver.find_element_by_xpath('//*[@id="sch_stat"]/option[2]').click()

    driver.find_element_by_xpath('//*[@id="wrap"]/div[3]/div[1]/form/div[2]/a').click()
    time.sleep(20)

    html_source = driver.page_source
    soup = BeautifulSoup(html_source, 'html.parser') # 페이지 파싱
    
    tt = now.strftime('%Y.%m.%d %H시 %M분')
    print(tt)
    print('############',t,'시작 ############')
          
    name = []
    chargn = []
    addr = []
    startc = []
    endc = []
    statdate = []

    for i in range(1,1244): # 갯수 달라짐 ㅠㅠ
        try:
            xpath = '//*[@id="history_span"]/tr[{}'.format(i)+']/td[2]/a'
            driver.find_element_by_xpath(xpath).click()

            html_source = driver.page_source
            soup = BeautifulSoup(html_source, 'html.parser')

            name.append(soup.find(id="detail_statNm").text)
            chargn.append(soup.find(id="detail_chgerId").text)
            addr.append(soup.find(id="detail_addrDoro").text)
            startc.append(soup.find(id="detail_lastTsdt").text)
            endc.append(soup.find(id="detail_lastTedt").text)
            statdate.append(soup.find(id="detail_statUpdDt").text)
            print('####',i,'완료 ####')
            
        except Exception:
            pass
          
    df = pd.DataFrame(None)
    df['충전소명'] = name
    df['충전기ID'] = chargn
    df['주소'] = addr
    df['마지막충전시작일시'] = startc
    df['마지막충전종료일시'] = endc
    df['상태갱신일시'] = statdate
    
    #tt = now.strftime('%Y-%m-%d %H:%M:%S')
    df.to_excel('충전일시_{}'.format(tt)+'.xlsx')
        
    time.sleep(1800)
```
